```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.describe()
```

```python
df.isnull().sum() * 100 / len(df)
```

```python
df.nunique()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df.drop(columns= 'deck', inplace=True)
```

```python
df.columns
```

```python
sns.countplot(data=df, x='survived')
```

```python
sns.barplot(data=df, x='survived', y= 'age', hue='who')
```

```python
sns.histplot(data=df, x='age', hue='who')
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='pclass')
plt.ylim(0, 300)
```

```python
sns.pairplot(data=df, hue='survived')
```

```python
df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
import seaborn as sns

sns.lmplot(data=df[df['fare'] < 300],x='age',y='fare',hue='who',col='sex')
```

```python
# how to add a culomn and make a  age group
df.columns
```

```python
df['age_group']= pd.cut(df['age'], [0,10,18,30,50,80],
  labels= ['kids', 'child', 'youth', 'midel_age', 'old']
                        )
```

```python
df.head()
```

```python
df.groupby('age_group')['survived'].mean().plot(kind='bar')
```

```python
df['family_size'] = df['sibsp'] + df['parch']

df['family_group'] = pd.cut(
    df['family_size'],
    bins=[0,1,4,20],  # ranges: 1 alone, 2-4 small family, 5+ large family
    labels=['Alone', 'Small_Family', 'Large_Family']
)
```

```python
df['family_size']= df['sibsp'] + df['parch']


df['family_size']= pd.cut(df['family_size'], bins=6)
```

```python
df.columns
```

```python
df.groupby('family_size')['survived'].mean().plot(kind='bar')
```
